Separo la fecha de los min5 y elimino la columna de 'Timestamp' y 'Volume Currency'

```python
times = [i.split() for i in data['Timestamp']]

date = []
time = []
for i in times:
    date.append(i[0])
    time.append(i[1])

data_clean = data.drop(['Timestamp'], axis=1)
data_clean.insert(0, "Date", date, False) 
data_clean.insert(1, '5min', time, False)

display(data_clean.head())
```

```python
data_clean['Open'].replace('—', 0, inplace=True)
data_clean['Close'].replace('—', 0, inplace=True)
```

```python
data_clean['Close'] = data_clean['Close'].astype(float)
data_clean['Open'] = data_clean['Open'].astype(float)
```
